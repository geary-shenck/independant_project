```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dfa = pd.read_csv("/Users/geary.shenck/Downloads/leafly_strain_data/leafly_strain_data.csv")
for col in dfa.iloc[:,6:].columns:
    dfa[col] = dfa[col].str.strip("%").astype(int)/100
    #if dfa[col].sum() < 5:
    #    dfa.drop(columns=col,inplace=True)
dfa
```

```python
dfa.iloc[:,6:].replace(0,np.nan)
```

```python
dfa["most_common_terpene"].unique().tolist()
```

```python
mean_terps={}
for each in dfa["most_common_terpene"].unique().tolist():
    mean_terps = (dfa[dfa["most_common_terpene"] == each].iloc[:,6:].mean()).to_dict(each)
mean_terps
```

```python
plt.figure(figsize=(20, 30))
for i, col in enumerate(dfa.iloc[:,6:].columns.tolist()): # List of columns
        plot_number = i + 1 # i starts at 0, but plot nos should start at 1
        plt.subplot(8,8, plot_number) # Create subplot.
        plt.title(col) # Title with column name.
        dfa[dfa[col]>0][col].hist(bins=10) # Display histogram for column.
        plt.grid(False) # Hide gridlines
```
